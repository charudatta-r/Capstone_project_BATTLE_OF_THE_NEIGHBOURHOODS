## THE BATTLE OF THE NEIGHBOURHOODS

### 1. INTRODUCTION & BUSINESS PROBLEM:

#### 1.1. PROBLEM BACKGROUND:

The aim of this project is to find a safe and secure location for opening of Grocery shop in New York. The New York City often called New York (NY), is the most populous city in the United States. City is the centre of the New York metropolitan area, the largest metropolitan area in the world by urban landmass. It is diverse and is the financial capital of not only USA but also of the world. New York has emerged as a global node of creativity and entrepreneurship. It provides lot of business opportunities and business friendly environment. As many as 800 languages are spoken in New York, making it the most linguistically diverse city in the world (Turin, n.d.). It has attracted many different players into the market. It is a global hub of business and commerce. New York is major capital for any business in US.

As a result, market in City is highly competitive. If you are looking to start a business, then cost will be one of the highest. Thus, market analysis is very important for any business starting new setup or expanding existing business. After the analyses, the insights getting from it will give good understanding to entrepreneur and will help them to derive their strategy to tap the market.
